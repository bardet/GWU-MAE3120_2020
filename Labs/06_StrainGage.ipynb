## Introduction and Background

Strain gages are some of the most frequently used resistive sensors. They feature a metal foil mounted on a thin epoxy, which serves as the support material. Because they are manufactured using the same techniques as for printed circuits, they can be made into a number of different configurations based on the needs of the experiment. These include single and multi-dimensional strain gages such as rosettes, which contain multiple elements and measure the strain at different angles. In addition, strain gages are also very light and flexible so they do not affect the performance of the system nor do they need to be mounted to a flat surface. The surface does however; need to be clean and smooth.

The principle behind strain gages is very simple. As a wire changes length and cross-sectional area, the resistance in turn also changes according to the equation below:

$$R = \frac{\rho L}{A}$$

Where $R$ is the resistance, $\rho$ is the resistivity of the wire, $L$ its length, and $A$ its cross-sectional area. Since the strain gage is bonded to the strained member, as the member deforms so will the strain gage. This in turn produces changes to the length and cross-sectional area, which change the resistance and allow the strain to be measured.

To correlate the strain of a given member to the change in resistance of the gage, there are additional dependencies on the material of the metal foil, the change in resistivity as the member strains, and Poisson's ratio, which need to be considered. If the strain remains in the elastic range, then these dependencies will remain almost constant. Manufacturer's of strain gages supply what is termed a gage factor ($S$) for each gage which is determined by the equation below for a wire with a square cross-section, where $\nu$ is Poisson's ratio, and $\epsilon_a = \delta L\,/\,L$ is the axial strain. 

$$S = 1 + 2 \nu + \frac{\delta \rho\,/\,\rho}{\epsilon_a} = \frac{\delta R\,/\,R}{\epsilon_a}$$

Finally, in the elastic regime, axial strain, $\sigma_a$, and stress are related with Hooke’s law:

$$\sigma_a = \frac{F}{A} = E \epsilon_a$$

With $F$ the axial load, and $E$ the Young elastic modulus of the material under test.

Below are some properties of the strain gage used for this lab:

- $R_{nominal}$: $350\,\Omega$

- $S$: $2.105$

- **Material**: Constant

To measure strain there are several configurations for strain gage mounting: the quarter, half, and full bridges. Quarter bridges are nonlinear with respect to strain, but with a half or full bridge, it is linear with respect to strain and the sensitivity is increased because several gages are in compression and in tension. In general, the change in resistance is very small, thus the output still needs to be amplified to obtain an easily measurable signal. ### Part I - Wheatstone Bridge

<img src="img/Wheatstone.png">

The bridge configuration is made of two voltages dividers, one for each of the legs. The difference in the voltages gives the change in resistance and can be calibrated to the physical quantities that are being measured. The Wheatstone bridge is used in many configurations such as a quarter, half, or full bridge. In the case of the quarter bridge only one resistance is unknown and the other three resistance values are matched as closely as possible to the nominal resistance value. In the case of the half bridge two resistance values are unknown and the other resistors are chosen to be as close to the nominal resistance values as possible, and in the full bridge all four resistance values are unknown.

When all the resistances are equal, the bridge is “balanced” and the output is $V_o$ = 0V. Due to tolerances on the electrical components it is difficult to obtain exactly 0V for the balanced bridge, however the value will be very close and this value recorded and later subtracted from the measurements. The output voltage of a Wheatstone bridge is given by, with $V_s$ the supply voltage:

$$V_o = V_s \frac{R_1 R_3 - R_2 R_4}{(R_2 + R_3)(R_1 + R_4)}$$

Assuming than the initial values of the resistances are such that the bridge is balanced, and that the value of the resistances is affected by a small amount ($R_i = R_{i,i} + \delta R_i$, with $\delta R_i$ the resistance variation and $R_{i,i} \gg \delta R_i$), at first order in $\delta R$ and for $R_{1,i} = R_{2,i} = R_{3,i} = R_{4,i}$, the output voltage of the bridge is affected by changes in resistance in the following way:

$$\frac{V_o}{V_s} = \frac{R_{2,i} + R_{3,i}}{(R_{2,i} + R_{3,i})^2} \left(\frac{\delta R_1}{R_{1,i}} - \frac{\delta R_2}{R_{2,i}} + \frac{\delta R_3}{R_{3,i}} - \frac{\delta R_4}{R_{4,i}}\right)$$
